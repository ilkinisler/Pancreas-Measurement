```python
import SimpleITK as sitk
import os
from os.path import join

def reorient_images(src_dir, target_dir):
    for img_name in os.listdir(src_dir):
        if img_name.startswith('.'): continue
        img_path = join(src_dir, img_name)
        stik_img = sitk.ReadImage(img_path, sitk.sitkFloat32)
        
        reoriented = sitk.DICOMOrient(stik_img, 'PLS')
        #print(img_name, stik_img.GetDirection())
        print(img_name, reoriented.GetDirection())

        # Save the reoriented image in the target directory
        target_file_path = join(target_dir, img_name)
        sitk.WriteImage(reoriented, target_file_path)

        print(f"Reoriented and saved: {img_name}")
```

```python
#LABELS

def main():
    centers = ['mayo_arizona', 'nyu', 'allegheny']
    for center in centers:
        src_dir = f'/data/ayc9699/dataset/pancreas_ipmn_ilkin/cyst_segmentation/labels/raw/{center}'
        target_dir = f'/data/ayc9699/dataset/pancreas_ipmn_ilkin/cyst_segmentation/labels/reoriented/{center}_PLS'
        reorient_images(src_dir, target_dir)

if __name__ == '__main__':
    main()
```

```python
#IMAGES

def main():
    centers = ['mayo_arizona', 'nyu', 'allegheny']
    for center in centers:
        src_dir = f'/data/ayc9699/dataset/pancreas_ipmn_ilkin/{center}/reoriented'
        target_dir = f'/data/ayc9699/dataset/pancreas_ipmn_ilkin/cyst_segmentation/images/reoriented/{center}_PLS'
        reorient_images(src_dir, target_dir)

if __name__ == '__main__':
    main()
```

```python
import os
def replace_double_underscores(folder_path):
    for filename in os.listdir(folder_path):
        if '__' in filename:
            new_filename = filename.replace('__', '_')
            old_file_path = os.path.join(folder_path, filename)
            new_file_path = os.path.join(folder_path, new_filename)
            os.rename(old_file_path, new_file_path)
            print(f'Renamed: {filename} -> {new_filename}')

if __name__ == "__main__":
    folder_path = '/data/ayc9699/dataset/pancreas_ipmn_ilkin/cyst_segmentation/labels/reoriented/mayo_arizona_PLS'
    replace_double_underscores(folder_path)
```
